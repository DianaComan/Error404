Prelucrarea Datelor spre a fi folosite in sistem( curatarea, generalizarea, maparea datelor)

```
#Prelucrarea Datelor


df_train = prelucrareDate("Date antrenare.csv")
df_test = prelucrareDate("Date test.csv")
df_bonus = prelucrareDate("Date bonus.csv")
```

Encoding-ul datelor din Dataframe-uri pentru a putea fi procesate de algoritmul de regresie

```
df_train = encodingData(df_train)
df_test = encodingData(df_test)
df_bonus = encodingData(df_bonus)
```

Preluarea Datelor de Train/Test si antrenarea modelului de Regresie Logistica

```
X_train = df_train[['confirmare contact cu o persoană infectată','grup dată rezultat', "tuse", "temp" , "frisoane",'seaca', 'musculare']]
X_test = df_test[['confirmare contact cu o persoană infectată', 'grup dată rezultat', "tuse", "temp" , "frisoane",'seaca', 'musculare']]
X_bonus = df_bonus[['confirmare contact cu o persoană infectată', 'grup dată rezultat', "tuse", "temp" , "frisoane",'seaca', 'musculare']]

#febra, tuse, temp, etc - sunt simptome particulare


y_train = df_train['rezultat']
y_test = df_test['rezultat']
y_bonus = df_bonus['rezultat']

lR = LogisticRegression()

lR.fit(X_train, y_train)

y_pred = lR.predict(X_test)
y_pred_inf = lR.predict(X_bonus)
```

Calcularea Metricilor

```
performanceMetrix(y_pred, y_test)
```

Calcularea Inferentei

```
performanceMetrix(y_pred_inf, y_bonus)
```
